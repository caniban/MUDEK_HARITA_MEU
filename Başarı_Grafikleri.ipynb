```python
import matplotlib.pyplot as plt
import numpy as np

raw_data_exam1_scores_25 = """AYŞE GÜL 78
MUHAMMET FURKAN EKİCİ 81
NUMAN ÖZBEK 83
İBRAHİM HALİL ARSLAN 81
EFEKAN YAVUZ 78
MUHAMMET ÇAKMAK 81
İSMAİL KILINÇ 78
LOUNA JASRI 50
TRTELL AMINO 98
SİYAHAN SAKMAN 98
BARIŞ ÖLMEZ 98"""


ogrenci_verileri_25 = {}

# Parse the initial data
for line in raw_data_exam1_scores_25.strip().split('\n'):
    # Strip each line to remove any leading/trailing whitespace, including extra spaces after the score
    parts = line.strip().rsplit(' ', 1)
    name = parts[0]
    score = int(parts[1])
    ogrenci_verileri_25[name] = score



ogrenci_yuzdeleri = {isim: notu for isim, notu in ogrenci_verileri_25.items()}



# Convert to lists for plotting
ogrenci_isimleri = list(ogrenci_yuzdeleri.keys())
yuzdesel_notlar = list(ogrenci_yuzdeleri.values())

# Function to determine colors based on score ranges (unchanged)
def get_color(score):
    if 0 <= score < 50:
        return 'darkred'  # Insufficient
    elif 50 <= score <= 60:
        return 'orange'   # Needs Improvement
    elif 60 < score <= 70:
        return 'gold'    # Sufficient
    elif 70 < score <= 90:
        return 'forestgreen'  # Good
    elif 90 < score <= 100:
        return 'blue'    # Excellent
    else:
        return 'grey'    # Undefined/error case

# Assign colors based on scores
renkler = [get_color(notu) for notu in yuzdesel_notlar]

# Create the plot and set its size
plt.figure(figsize=(14, 8))
bars = plt.bar(ogrenci_isimleri, yuzdesel_notlar, color=renkler)

# Set title and axis labels
plt.xlabel("Öğrenci Adı", fontsize=12)
plt.ylabel("Başarı Yüzdesi (%)", fontsize=12)
plt.title("Öğrenim Çıktısı 1 Başarı Grafiği", fontsize=16, fontweight='bold')

# Set y-axis limits and ticks, with extra space at the top
plt.ylim(0, 105)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(np.arange(0, 101, 10), fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add percentage labels above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{round(yval)}%', ha='center', va='bottom', fontsize=9)

# Create legend (unchanged)
handles = [
    plt.Rectangle((0,0),1,1, color='darkred', label='0-50 (Yetersiz)'),
    plt.Rectangle((0,0),1,1, color='orange', label='50-60 (İyileştirilebilir)'),
    plt.Rectangle((0,0),1,1, color='gold', label='60-70 (Yeterli)'),
    plt.Rectangle((0,0),1,1, color='forestgreen', label='70-90 (İyi)'),
    plt.Rectangle((0,0),1,1, color='blue', label='90-100 (Mükemmel)')
]
plt.legend(handles=handles, title="Başarı Kriterleri", loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

# Adjust layout to make room for the legend
plt.tight_layout(rect=[0, 0, 0.85, 1])

# Save the plot as a high-resolution PNG
plt.savefig("Veri Bilimi Öğrenim Çıktısı 1_Kriterli.png", dpi=300)
plt.show()
```
